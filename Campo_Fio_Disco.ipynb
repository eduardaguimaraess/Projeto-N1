```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Constante de Coulomb
K = 8.99 * (10**9)

# Definição dos parâmetros do fio
L = 1.0        # comprimento do fio (m)
lam = 1e-6     # densidade linear de carga (C/m)

# Função: campo elétrico do fio finito no ponto (0,y)
def campoFio(y):
    Ey = (2 * K * lam * L) / (y * math.sqrt(L**2 + y**2))
    return Ey

# Função: campo elétrico de uma carga pontual equivalente Q = 2λL
def campoPontual(y):
    Q = 2 * lam * L
    E = K * Q / (y**2)
    return E

# Valores de y
y_vals = np.logspace(-4, 4, 400)

# Calcula os campos
E_fio_vals = [campoFio(y) for y in y_vals]
E_pontual_vals = [campoPontual(y) for y in y_vals]

# Exemplo de saída numérica
y_exemplo = 1.0
print("Distância y =", y_exemplo, "m")
print("Campo do fio finito (N/C):", campoFio(y_exemplo))
print("Campo da carga pontual equivalente (N/C):", campoPontual(y_exemplo))

# Gráfico
plt.figure(figsize=(8,5))
plt.plot(y_vals, E_fio_vals, label="Fio finito", linewidth=2)
plt.plot(y_vals, E_pontual_vals, "--", label="Carga pontual equivalente", linewidth=2)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Distância y (m) [escala log]")
plt.ylabel("Campo elétrico (N/C) [escala log]")
plt.title("Campo elétrico: fio finito vs carga pontual equivalente")
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Constante de Coulomb
K = 8.99 * (10**9)

# Definição de parâmetros do disco
raioDisco = 1.0                # raio do disco (m)
densidadeSuperficial = 1e-6    # densidade superficial de carga (C/m^2)

# Função: campo elétrico do disco a uma distância y
def campoDisco(y):
    # Usando a relação K = 1 / (4 pi eps0)
    eps0 = 1 / (4 * math.pi * K)
    Ey = (densidadeSuperficial / (2 * eps0)) * (1 - (y / math.sqrt(y**2 + raioDisco**2)))
    return Ey

# Valores de y de 10^-4 a 10^2
yValores = np.logspace(-4, 2, 400)
EValores = [campoDisco(y) for y in yValores]

# Exemplo numérico
yExemplo = 0.5
print("Distância y =", yExemplo, "m")
print("Campo elétrico no ponto P (N/C):", campoDisco(yExemplo))

# Gráfico
plt.figure(figsize=(8,5))
plt.plot(yValores, EValores, label="Campo elétrico do disco", linewidth=2)
plt.xscale("log")
plt.xlabel("Distância y (m) [escala log]")
plt.ylabel("Campo elétrico (N/C)")
plt.title("Campo elétrico no eixo de um disco carregado")
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```
